```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import pairwise_distances_argmin_min

import seaborn as sns

plt.rcParams["figure.figsize"] = (8,5)
```

# Ejemplo 1

```python
# Importar datos
data = datasets.load_iris()
X = data["data"]
y = data["target"]
colnames = ["A", "B", "C", "D"]
```

```python
df = pd.DataFrame(X, columns=colnames )
```

```python
Nc = range(1,8)
kmeans = []
for i in Nc:
    kmeans.append(KMeans(n_clusters=i))
    
score = []
for i in range(len(Nc)):
    score.append(kmeans[i].fit(X).score(X))
    
plt.plot(Nc, score)
plt.xlabel("Número de clústeres")
plt.ylabel("Puntuación")
plt.title(" Curva de codo");
```

```python
# Graficar los puntos

kmeans = KMeans(n_clusters=3).fit(X)
C = kmeans.cluster_centers_
pred = kmeans.predict(X)

col = ["red","blue","green"]

asig = []
for i in pred:
    asig.append(col[i])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2],c=asig)
ax.scatter(C[:,0],C[:,1],C[:,2],marker="*",c=col,s=1000);
```

```python
df["grupo"] = pred
sns.scatterplot(x="B", y="C", hue="grupo", data=df, palette="viridis")
```

```python
df.groupby(by="grupo").count()
```

```python
x1 = X[:,0]
x2 = X[:,1]
x3 = X[:,2]
x4 = X[:,3]

plt.scatter(x2,x3,c=pred);
```

```python
plt.scatter(x2,x3,c=y)
```

## Ejemplo 2

```python
dataframe = pd.read_csv("vivienda_clasif.csv",sep=";", decimal=".")
label = dataframe["Clasifi"]
X = dataframe.drop(["Clasifi","habitaciones"],1)

X.describe()
```

```python
X.hist();
```

```python
# Tomar los valores como un array

X = np.array(X)
X.shape
```

```python
# Obtener el valor de k

Nc =range(1,15)
kmeans = []
for i in Nc:
    kmeans.append(KMeans(n_clusters=i))

score = []
for i in range(len(Nc)):
    score.append(kmeans[i].fit(X).score(X))

plt.plot(Nc,score)
plt.xlabel("Número de clústeres")
plt.ylabel("Puntuación")
plt.title(" Curva de codo");
```

## Ejecutando el algoritmo de k-means para 5 clústeres

```python
kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print("Centroides: \n", centroids)
```

```python
# pronóstico de los clúster
pred = kmeans.predict(X)

# Centroides
colors = ['red','green','blue','cyan','yellow']
col = []
for i in pred:
    col.append(colors[i])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2],c=col)
ax.scatter(centroids[:,0],centroids[:,1],centroids[:,2],marker="*",c=colors,s=500);
```

# Ejemplo 3

```python
data = pd.read_csv("shopping-data.csv",sep=",")
```

```python
data = pd.get_dummies(data)
data = data.drop(["CustomerID"],axis=1)
nombres = data.columns
```

```python
data[data["Age"]==19]
```

```python
data = np.asarray(data)
```

```python
# Obtener el valor de k

Nc =range(1,30)
kmeans = []
for i in Nc:
    kmeans.append(KMeans(n_clusters=i))

score = []
for i in range(len(Nc)):
    score.append(kmeans[i].fit(data).score(data))

plt.plot(Nc,score)
plt.xlabel("Número de clústeres")
plt.ylabel("Puntuación")
plt.title(" Curva de codo");
```

```python
kmeans = KMeans(n_clusters=6).fit(data)
centroids = kmeans.cluster_centers_
#print("Centroides: \n", centroids)


# pronóstico de los clúster
pred = kmeans.predict(data)

```

```python
# Centroides
colors = ['red','green','blue','cyan','yellow',"lightblue"]
col = []
for i in pred:
    col.append(colors[i])

fig = plt.figure(figsize=(8,6))
plt.scatter(data[:,1],data[:,2], c=col);
```
